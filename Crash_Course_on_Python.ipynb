<h2 style="font-size: 20px; font-weight: bold;">Python Variable Naming Cheat Sheet</h2>

** Use descriptive names**  
Choose names that clearly describe what the variable stores.  
 Example: `total_payment`  
 Example: `x`  

---

** Start with a letter**  
Variable names must begin with a letter (az or AZ), not a number.  
 Example: `user1`  
 Example: `1user`  

---

** Don't use spaces**  
Spaces are not allowed. Use underscores `_` to separate words.  
 Example: `user_name`  
 Example: `user name`  

---

** Python is case-sensitive**  
`age`, `Age`, and `AGE` are different variables. Pay attention to capitalization.  
 Example: `is_active`  
 Confusing `is_active` with `Is_active` can cause bugs.  

---

** Use meaningful names**  
Avoid vague letters or unclear abbreviations.  
 Example: `body_temperature`  
 Example: `t`  

---

** Keep names concise but clear**  
Avoid overly short or extremely long names.  
 Example: `invoice_total`  
 Example: `it`, `total_amount_to_be_paid_for_the_entire_month_invoice`  

---

** Avoid reusing variable names**  
Reusing names can overwrite data and lead to bugs.  
 Use unique names for different purposes  
 Dont use `result` repeatedly in different contexts

<h2 style="font-size: 20px; font-weight: bold;">Python Operators Cheat Sheet</h2>

** Arithmetic Operators**  
Used to perform basic mathematical operations.  

- `+` Addition: Adds two values  
   Example: `3 + 2`  `5`

- `-` Subtraction: Subtracts the second value from the first  
   Example: `5 - 2`  `3`

- `*` Multiplication: Multiplies two values  
   Example: `4 * 3`  `12`

- `/` Division: Divides the first value by the second (returns float)  
   Example: `10 / 4`  `2.5`

- `**` Exponentiation: Raises a number to the power of another  
   Example: `2 ** 3`  `8`

- `%` Modulus: Returns the remainder of a division  
   Example: `10 % 3`  `1`

- `//` Floor Division: Returns the integer part of a division  
   Example: `10 // 3`  `3`

---

** Comparison Operators**  
Used to compare values and return Boolean results.  

- `==` Equal to: Checks if two values are equal  
   Example: `3 == 3`  `True`

- `!=` Not equal to: Checks if two values are different  
   Example: `5 != 3`  `True`

- `>` Greater than: Checks if left value is greater  
   Example: `5 > 3`  `True`

- `<` Less than: Checks if left value is smaller  
   Example: `2 < 4`  `True`

- `>=` Greater than or equal to: Checks if left value is greater or equal  
   Example: `5 >= 5`  `True`

- `<=` Less than or equal to: Checks if left value is smaller or equal  
   Example: `3 <= 4`  `True`

---

### Examples:

````python

print(32 == 30+2)   # The == operator checks if the 2 values are
                    # equal to each other. If they are equal,
                    # Python returns a True result.

print(9/3 != 3*1)   # In this last example, 9/3 != 3*1 (or 3 != 3)
                    # is false. So, Python returns a False value.

print(4/2 > 8-4)    # If the > operator finds that the left value
                    # is NOT greater than the right value, the
                    # comparison will return a False result.

print(4/2 < 8-4)    # The < operator checks  if the left value is
                    # less than the right side. If true, the
                    # comparison returns a True result.

print(18/2 >= 15)   # If the >= comparison determines that the left
                    # value is NOT greater than or equal to the
                    # right, it returns a False result.

print(12*2 <= 30)   # The <= operator checks if the left value is
                    # less than or equal to the right value. In
                    # this case, the left value is less than the
                    # right value. Again, if one of the two
                    # conditions is true, Python returns a True
                    # result.
````

<h2 style="font-size: 20px; font-weight: bold;">Python Logical Operators Cheat Sheet</h2>

** and (Logical AND)**  
Both conditions must be `True` for the result to be `True`.  
 Example: `(5 > 1 and 5 < 10)`  `True`  
 Example: `(5 > 1 and 5 > 10)`  `False`

---

** or (Logical OR)**  
Only one condition needs to be `True` for the result to be `True`.  
 Example: `(color == "blue" or color == "green")`  `True` *(if `color = "blue"`)*  
 Example: `(5 < 2 or 3 == 3)`  `True`

---

** not (Logical NOT)**  
Inverts the result: turns `True` into `False` and vice versa.  
 Example: `(not "A" == "A")`  `False`  
 Example: `(not 5 > 10)`  `True`

---


### Examples:

````python
print((6*3 >= 18) and (9+9 <= 36/2))     # True and True = True

print((15/3 < 2+4) or (0 >= 6-7))        # True or True = True

print(16 <= 4**2 or 9**(0.5) != 3)       # True or False = True

today = "Monday"
print(not today == "Tuesday")

# The "today" variable states today is Monday. This makes the comparison
# "today == Tuesday" False. The logical operator "not" inverts the False
# result to become True. In other words, this expression asks if it is
# false that today is not Tuesday. More succinctly, "not False" means
# True."
````

## 1.4 Conditional Statements

Suppose a customer is planning to buy a mobile phone but has a limited budget. Thus, his decision to buy is based on the condition that the price comes under his budget. Let's say his budget is $600.

Q. Write a code in Python that prints whether the customer can buy the iPhone or not based on his budget.

Sometimes we want to execute a block of code based on a particular condition.
In these types of situations, we write conditional statements.

The if-else statement in Python can be used for decision-making. The general syntax of an if-else statement is:

```
if (test expression):
    Body of if
else:
    Body of else
```

Python is a tabbed based language, so we need to be VERY careful about tabs!!!


````python
# define the budget price
budget = int(input('Enter your budget(in dollars): '))

# if-else statement
if price <= budget:
  print('Congrats! You can buy the Iphone')
else:
  print('Sorry! The mobile price is more than your budget')
````

Python also supports a shorthand for `if-else` using a ternary operator (contidional expression), which allows you to write a single line conditional.
```
value_if_true if condition else value_if_false
```

````python
x = 5
message = "x is even" if x % 2 == 0 else "x is odd"
print(message)
````

If there are multiple conditions, we may use another conditional statement called if-elif-else statement. The general syntax of the if-elif-else statement is:

```
if (test expression):
    Body of if
elif (test expression):
    Body of elif
else:
    Body of else
```



Q. Suppose the price of the 32 GB, 64 GB, and 128 GB iPhones are \$600, \$700, and \$900 respectively. Write a conditional statement to print the price based on the internal storage of the phone.

````python
Memory = int(input('Enter the memory: '))

if Memory == 32:
  print('The price of the phone is $600')
elif Memory == 64:
  print('The price of the phone is $700')
elif Memory ==128:
  print('The price of the phone is $900')
else:
  print('Please enter a valid memory requirement')
````

## 1.6 List Comprehensions

MobiWorld has decided to provide an instant discount of 5% on all their products during the upcoming sale. We want to check the discounted prices of the products.

We already have the `price_list` with us. Let's use it to get the discounted prices.

One way of creating a discounted price list is by looping over the elements of `price_list`, subtracting the discount from the price, and adding it one at a time to a new list `discounted_price_list`.

````python
discounted_price_list=[]

for x in price_list:
  discounted_price = x - (x*(5/100))
  discounted_price_list.append(discounted_price)

print(discounted_price_list)
````

An equivalent way of obtaining the same result is by using **list comprehensions**.

* List comprehension is a type of control structure for creating a list from an existing iterable (like tuples, strings, arrays, lists, etc).
* They offer a shorter and more appealing syntax and are often faster than explicit for loops in creating a list from an existing iterable.

````python
discounted_price_list = [x - (x*(5/100)) for x in price_list]
print(discounted_price_list)
````

Let's breakdown the syntax of the above list comprehension. There are three parts:
1. `[ ]`: The outer square brackets tells that we are creating a list
2. `x-(x*(5/100)`: The expression to evaluate or the computation to perform
3. `for x in price_list`: The iterable from which elements (x) will be fetched

Customers often want to check whether any of the discounted prices during the sale are within their budget.

Let's help the store display the same based on the budget mentioned by the customer by comparing the discounted price with the budget for each of the four mobile phone prices.

````python
# asking for customer's budget
budget = int(input('Enter your budget(in dollars): '))

# creating a list of Yes/No based on budget and discounted prices
within_budget = ['Yes' if x <= budget else 'No' for x in discounted_price_list]
print(within_budget)
````

To get more clarity on the way the list comprehension syntax is formulated, we can expand the list comprehension to the usual for and if statements and think of this in a reversed way.

````python
within_budget = []

for x in discounted_price_list:
  if x <= budget:
    within_budget.append('Yes')
  else:
    within_budget.append('No')

print(within_budget)
````

````python
def format_address(address_string):

    house_number = ""
    street_name = ""

    # Separate the house number from the street name.
    address_parts = address_string.split()

    for address_part in address_parts:
       # Complete the if-statement with a string method.
       if address_part.isdigit():
           house_number = address_part
       else:
           street_name += address_part + " "

    # Remove the extra space at the end of the last "street_name".
    street_name = street_name.strip()

    # Use a string method to return the required formatted string.
    return "House number {} on a street named {}".format(house_number, street_name)

print(format_address("123 Main Street"))
# Should print: "House number 123 on a street named Main Street"

print(format_address("1001 1st Ave"))
# Should print: "House number 1001 on a street named 1st Ave"

print(format_address("55 North Center Drive"))
# Should print "House number 55 on a street named North Center Drive"
````
